# Importing 

```python
# Importing the Pandas library to work with dataframes
import pandas as pd

# Importing the Matplotlib library to plot data 
import matplotlib.pyplot as plt
```

# Data Preparation

# 1. Write a python program to merge data from each month into one CSV and read in updated dataframe. 

```python
# Defining a list of months
data = ["January","February","March","April","May","June","July","August","September","October","November","December"]

# Creating an empty list to store the dataframes
initial = []

# looping through each mont and read in the corresponding CSV file
for i in data:

    to_add = "Sales_"+i+"_2019.csv" # constructing the filename for the current month

    add = pd.read_csv(to_add) # reading in the CSV file
    
    initial.append(add) # adding the dataframes to the list
    
    
# Connecting all of the dataframes into a single dataframe
final_frame = pd.concat(initial)


# Writing the final dataframe to a CSV file
final_frame.to_csv("final.csv", index = False)


# Reading in the final dataframe from the CSV file
final_frame = pd.read_csv("final.csv")


final_frame
    
```

# 2. Write a python program to remove the NaN missing values from updated dataframe. 

```python
# Dropping any rows containing NaN(not a number) values from the final dataframe
removed_NaN = final_frame.dropna()

removed_NaN
```

# 3. Write a python program to convert Quantity Ordered and Price Each to numeric.

```python
# Getting the Quantity Ordered and Price Each columns from the dataframe
qty = removed_NaN["Quantity Ordered"]
price = removed_NaN["Price Each"]

# Converting the Quantity Ordered and Price Each columns into numeric types
qty = pd.to_numeric(qty)
price = pd.to_numeric(price)


print("Quantity data type =",qty.dtype)
print("Price data type =",price.dtype)
```

# 4. Create a new column named Month from Ordered Date of updated dataframe and convert it to integer as data type.

```python
#Getting the Order Date column from the dataframe 
ord_date = removed_NaN["Order Date"]
 
# Converting the Order Date column to datetime
ord_date = pd.to_datetime(ord_date)
 

# Adding a new column to the dataframe named Month which contain the month number from the Order date column
removed_NaN["Month"] = ord_date.dt.month.astype(int)


removed_NaN
```

# 5. Create a new column named City from Purchase Address based on the value in updated dataframe.

```python
# Getting the Purchase Address column from the dataframe 
p_add = removed_NaN["Purchase Address"]

# Splitting the Purchase Address column by comma and get the second element, which is city name
removed_NaN["City"] = p_add.str.split(",").str[1]

removed_NaN
```

# Data Analysis

# 1. Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable. 

```python
# Choosing Price Each column in a variable price
price = removed_NaN["Price Each"]

# Calculating sum of Price Each column
Sum = price.sum()

print("The sum of Price Each column =",Sum)


#Calculating mean of Price Each column
mean = price.mean()

print("The mean of Price Each column =",mean)


#Calculating Standard Deviation of Price Each column
std_dev = price.std()

print("The Standard Deviation of Price Each column =",std_dev)


#Calculating skewness of Price Each column
sk_ns = price.skew()

print("The skewness of Price Each column =",sk_ns)


#Calculating kurtosis of Price Each column
kurts = price.kurtosis()

print("The kurtosis of Price Each column =",kurts)
```

# Data Exploration

# 1. Which Month has the best sales? and how much was the earning in that month? Make a bar graph of sales as well. 

```python
# Creating monthly sales by grouping data by month and calculating sum of the Quantity Ordered Column 
mth_sales = removed_NaN.groupby("Month")["Quantity Ordered"].sum()

# Finding the month with the best sales
best_mth = mth_sales.idxmax()

#conversion of a numerical value representing a month into its corresponding month name.
best_mth_word = pd.to_datetime(str(best_mth), format='%m').strftime('%B')

print("The Best month for sales =",best_mth_word)


# Finding the earining of the month which has best sales
high_sales = mth_sales.max()

print("The earning of the",best_mth_word,"month =",high_sales)


# Plotting in a bargraph of the monthly sales
plt.bar(mth_sales.index, mth_sales.values, color = "yellow", edgecolor = "green")
plt.title("Monthly sales")
plt.xlabel("Month")
plt.ylabel("Total sales")
plt.show()
```

# 2. Which city has sold the highest product?

```python
# Creating city sales by grouping data by City and calculating sum of the Quantity Ordered Column 
cty_sales = removed_NaN.groupby("City")["Quantity Ordered"].sum()


#Finding the city which has the highest sold product
high_sales = cty_sales.idxmax()

print("The city with highest sold product =",high_sales)
```

# 3. Which product was sold the most in overall? Illustrate it through bargraph.

```python
# Creating product sales by grouping data by Product and calculating sum of the Quantity Ordered Column 
prod_sales = removed_NaN.groupby("Product")["Quantity Ordered"].sum()

# Finding the most sold product 
most_sold = prod_sales.idxmax()

print("The most sold product =",most_sold)


# Plotting in a bargraph of the product sales in a year
plt.bar(prod_sales.index, prod_sales.values, color='yellow', edgecolor="green")
plt.xticks(rotation='vertical')
plt.title("Product Sales in a Year")
plt.xlabel("Product")
plt.ylabel("Quantity Ordered")
plt.show()
```
